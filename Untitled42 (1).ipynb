```python
from datasets import load_dataset
dataset = load_dataset("imdb")
print(dataset['train'][0])
```

```python
import pandas as pd
data_dict = {'text': ['I rented I AM CURIOUS-YELLOW...', '"I Am Curious: Yellow" is...', 'If only to avoid...', 'This film was probably...', 'Oh, brother...after hearing...'],
             'label': [0, 0, 0, 0, 0]}
df = pd.DataFrame(data_dict)
df.head()
```

Tokenization

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
from transformers import TrainingArguments, Trainer
import os

```

```python
output_path = os.path.join(os.path.expanduser("~"), "sentiment_project_results")
training_args = TrainingArguments(
    output_dir=output_path,
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=1,

)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
)

trainer.train()
```

```python
import numpy as np
!pip install evaluate
import evaluate
```

```python
accuracy_metric = evaluate.load("accuracy")
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy_metric.compute(predictions=predictions, references=labels)
trainer.compute_metrics = compute_metrics
final_results = trainer.evaluate()
print(f"======================================")
print(f"Final Accuracy: {final_results['eval_accuracy'] * 100:.2f}%")
print(f"======================================")
```

Test Model


```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, device=0)
```

```python
test_sentences = [
    "This is the best movie I have ever seen in my entire life!", #   (Positive)
    "I wasted two hours of my life on this garbage.",            #   (Negative)
    "The acting was great, but the story was incredibly boring." # جملة مختلطة
]
```

```python
results = classifier(test_sentences)
```

```python
for text, res in zip(test_sentences, results):
    sentiment = "Positive" if res['label'] == 'LABEL_1' else "Negative"
    score = res['score'] * 100
    print(f"Text: {text}\nResult: {sentiment} ({score:.2f}% confidence)\n")
```
